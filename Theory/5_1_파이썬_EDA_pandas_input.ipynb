## loc
* 행(index) 인덱싱
* .loc[행]
* .loc[행, 열]
* .loc[조건식, 열]


```python
# loc 로 하나의 행을 가져옵니다.
df.loc[0]#첫 번째
```

```python
# loc로 2개의 행을 가져와 봅니다.
df.loc[[0,1]]
```

```python
# loc로 행, 열을 가져와 봅니다.
df.loc[0,"name"]
```

```python
# loc로 여러개의 행, 열을 가져와 봅니다.
df.loc[[0,1], ["name", "origin"]]
```
