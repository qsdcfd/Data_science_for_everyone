#### 학습 목표

범주형 변수의 산점도를 표현하는 다양한 방법을 알아봅니다.

서브플룻을 그리는 catplot알아보기

<br> 

#### 핵심 키워드

scatterplot

stripplot

swarmplot

catplot

kind: bar, point, box, violin, strip, swarm

---

#### Point

범주형 변수의 산점도를 scatterplot으로 표현하면 같은 곳에 점이 여러 개 찍혀서 값을 제대로 표현하기 어렵기에  범주형 데이터의 산점도를 표현할 때는  stripplot 을 사용하면 점을 흩어지게 그리게 됩니다. 하지만 이 그래프 역시 데이터가 많으면 점이 겹쳐 보이게 되는데 이 단점을 보완한 시각화가 swarmplot 입니다.


<br>

#### Code로 보기

*scatterplot

```
# scatterplot 으로 범주형 변수 그리기
sns.scatterplot(data=df, x="origin", y="mpg")
```
 ![](https://images.velog.io/images/qsdcfd/post/785ed320-5379-438c-aa9b-aad89b9f9f27/image.png)
 
*stripplot

 ```
# stripplot
sns.stripplot(data=df, x="origin", y="mpg")
```
![](https://images.velog.io/images/qsdcfd/post/44d0e87c-601d-4f39-990f-86786cce26f1/image.png)

*swarmplot

```
# swarmplot
plt.figure(figsize=(10,4))
sns.swarmplot(data=df, x="origin", y="mpg", size=3)
```

![](https://images.velog.io/images/qsdcfd/post/a54143c2-9a01-4a7e-92d7-d0a09824aee3/image.png)

*catplot

catplot 은 범주형 데이터의 서브플롯을 그릴 때 사용하고 기본 값은 stripplot이며 아래 catplot 을 통해 kind 값을 지정해 bar, point, box, violin, strip, swarmplot을 시각화 할 수 있습니다.

![](https://images.velog.io/images/qsdcfd/post/857a848d-514b-43ef-bc84-dd897ab596f7/image.png)

<br/>

**code**

```
# catplot
sns.catplot(data= df, x = "origin", y="mpg")
```

![](https://images.velog.io/images/qsdcfd/post/f4ace60e-287c-42dd-ade3-5e3d882ab197/image.png)

```
# catplot 으로 boxplot그리기
# catplot
sns.catplot(data= df, x = "origin", y="mpg", col= "cylinders", col_wrap = 3, kind="box")
```

![](https://images.velog.io/images/qsdcfd/post/9228301f-f7d9-4258-85db-d86d04ceb8de/image.png)

```
# catplot 으로 violinplot그리기
# catplot
sns.catplot(data= df, x = "origin", y="mpg", col= "cylinders", col_wrap = 3, kind="violin")
```

![](https://images.velog.io/images/qsdcfd/post/83daae90-ca07-49b8-a189-4e732a35ef28/image.png)

```
# catplot 으로 countplot그리기
# catplot
sns.catplot(data= df, x = "origin", col= "cylinders", col_wrap = 3, kind="count")
```

![](https://images.velog.io/images/qsdcfd/post/7b295941-1d7f-4cd0-b33f-474d2c473ddc/image.png)

```
# catplot 으로 boxplot그리기
# df 전체를 넣고 시각화 하기
sns.catplot(data= df, kind="box", aspect=3)
```

![](https://images.velog.io/images/qsdcfd/post/35bb2ae2-4e43-4c11-9ae3-8d750fec5a01/image.png)

```
# catplot 으로 violinplot그리기
# df 전체를 넣고 시각화 하기
sns.catplot(data= df, kind="violin", aspect=3)
```

![](https://images.velog.io/images/qsdcfd/post/c93fda67-20d7-43b7-aef1-731e86cbfd62/image.png)
